## Cluster kategori (Countlot / barplot and Pie chart)

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python

import pandas as pd


bins = [0, 18, 60, 100]  
labels = ['Young', 'Adult', 'Old']  

df['Age_Category'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


print(df[['Age', 'Age_Category']])
```

```python

bins = [0, 50, 100, 200, 600]  
labels = ['Cheap', 'Moderate', 'Expensive', 'Premium'] 


df['Fare_Category'] = pd.cut(df['Fare'], bins=bins, labels=labels, right=False)

print(df[['Fare', 'Fare_Category']].head())
```

## Data Preprocessing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Memuatkan dataset Titanic yang telah dibersihkan dan encode
titanic_encode = pd.get_dummies(titanic_selection, drop_first=True)

# Memilih ciri-ciri (X) dan sasaran (y)
X = titanic_encode.drop('Survived', axis=1)  # Semua kolum kecuali 'Survived'
y = titanic_encode['Survived']  # Sasaran adalah 'Survived'

# Membahagikan data kepada set latihan dan set ujian (70% latihan, 30% ujian)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Melakukan scaling pada ciri-ciri numerik
scaler = StandardScaler()

# Latih scaler pada data latihan dan transformasi kedua-dua set latihan dan ujian
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Run Model

```python
# Inisialisasi model KNN
knn = KNeighborsClassifier(n_neighbors=5)  # Anda boleh cuba pelbagai nilai k (contoh: 3, 5, 7)

# Latih model KNN menggunakan data latihan yang telah diskalakan
knn.fit(X_train_scaled, y_train)
```
